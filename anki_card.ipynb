```python
import pandas as pd
txt_path = 'word.txt' # استبدلها بمسار ملفك
import pandas as pd
import re



with open(txt_path, "r", encoding="utf-8") as file:
    lines = file.readlines()


arabic_pattern = re.compile(r'[؀-ۿ]+')
english_pattern = re.compile(r'[A-Za-z]+')

data = []
for line in lines:
    arabic_words = " ".join(arabic_pattern.findall(line))
    english_words = " ".join(english_pattern.findall(line))
    data.append([arabic_words, english_words])


df_text = pd.DataFrame(data, columns=["Arabic", "English"])


df_text
```
